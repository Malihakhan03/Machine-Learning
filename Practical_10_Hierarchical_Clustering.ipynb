```python
#step1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
path = "/content/drive/MyDrive/Datasets/Mall_Customers.csv"
df = pd.read_csv(path)
df
```

```python
#only requied columns and only two column x have
#x = df.iloc[:,[3,4]].values
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
x
```

```python
#feature scaling since hierarchical clustering uses distance calculation
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#step5
plt.figure(figsize=(8,5))

dendrogram(linkage(x_scaled, method='ward'))

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn import metrics
#assume optimal clusters=5
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
clusters = hc.fit_predict(x_scaled)
```

```python
#step 7
df['Cluster'] = clusters
print(df.head())
```

```python
#step 8
plt.figure(figsize=(8,6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='Set1')

plt.title("Hierarchical Clustering of Mall Customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100")
plt.show()
```
