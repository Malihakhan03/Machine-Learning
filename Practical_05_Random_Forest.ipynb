```python
x = df.drop(columns=['y'])
y = df['y']
```

```python
#One-hot encode categorical variables
#get_dummies(x)--> each unique category in data is converted into a separate column and the...
#x= pd.get_dummies(x)
```

```python
x=pd.get_dummies(x,dtype=int)
```

```python
x.head()
```

```python
x.shape
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7,test_size=0.3,random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#fit the model on the training set using the decision tree classifier
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100,n_jobs=-1, max_depth=5, oob_score=True, random_state=42)
rf_classifier.fit(x_train, y_train)
```

```python
#prediction
y_pred = rf_classifier.predict(x_test)
```

```python
y_pred
```

```python
#Accuracy Score, classification Report and confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print('Confusion matrix')
cm=confusion_matrix(y_pred,y_test)
print(confusion_matrix(y_pred,y_test))
```

```python
print('Accuracy score')
print(accuracy_score(y_pred,y_test))
```

```python
#checking oob score
rf_classifier.oob_score_
```

```python
print('Classification report')
print(classification_report(y_pred,y_test))
```

**Hyperparameter tunning using grid search cv**

```python
#GridSearchCV#

rf = RandomForestClassifier(random_state=42, n_jobs=-1)
param_grid = {
    'max_depth': [2, 3, 5],
    'min_samples_leaf': [5,10,20,50],
    'n_estimators': [10,25,30,50]
}

from sklearn.model_selection import GridSearchCV
#Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=4, n_jobs=-1, verbose=1, scoring='accuracy')

grid_search.fit(x_train, y_train)
```

```python
grid_search.best_score_
```

```python
#print best parameter for GridSearchCV
print(grid_search.best_params_)
```
