**2. pandas-data handling and preprocessing **

```python
import pandas as pd
```

```python
data = {
    'Age': [25,30,35,40],
    'Salary' : [50000,60000,70000,80000],
    'Department' : ['HR','IT','Finance','Marketing']
}
df = pd.DataFrame(data)
print(df)

```

```python
print(df.shape)
print()
print(df.info())
```

```python

print(df.describe)
```

```python
print(df.tail())
print()
print(df.head())
```

```python
#data slection
print(df['Age'])
print()
print(df[['Salary','Department']])
```

```python
#Handling missing values
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
print(df)
```

```python
#encoding categorical data
df_encoded = pd.get_dummies(df,drop_first=True)
print(df_encoded)
```

```python
#3. matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,30,40]
```

```python
#line plot
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show()
```

```python
#bar plot
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Plot')
plt.show()
```

```python
#scatter plot
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

**4. Seaborn-Statistical visualizationt**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load sample dataset
data = sns.load_dataset("iris")
```

```python
#pair plot
sns.pairplot(data,hue='species')
plt.show()
```

```python
#Heatmap
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#box plot
sns.boxplot(x='species',y='sepal_length',data=data)
plt.show()
```

```python
#box plot
sns.boxplot(x='species',y='petal_length',data=data)
plt.show()
```
