```python
#only requied columns and only two column x have
x = df.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#feature scaling
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
inertia_values = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(x_scaled)
    inertia_values.append(kmeans.inertia_)

print(inertia_values)
#this step is to know the value of k by using elbow method
```

```python
plt.figure()
plt.plot(range(1, 11),inertia_values, marker='o')
plt.xlabel('Number of Clusters(x)')
plt.ylabel('Inertia')
plt.title("Elbow Method for optimal k")
plt.show()
#where drastically change that is consider for k value
```

```python
#step6: apply k-means(Assume k=5)
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(x_scaled)

# add cluster labels
df['Cluster'] = clusters
print(df.head())
```

```python
df
```

```python
#silhoutte score use to check quality(more towards 1 more optimal)
#step 7 evaluate clustering
print("inertia:", kmeans.inertia_)

print("Silhouette Score:", silhouette_score(x_scaled, clusters))
```

```python
#step8 visualize
plt.figure()
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='Set1')

#plot centroid
centroids = scaler.inverse_transform(kmeans.cluster_centers_)

plt.scatter(centroids[:,0],
            centroids[:,1],
            s=200,
            c='black',marker='X',label='Centroids')

plt.title("Customer segmentation using K-means")
plt.legend()
plt.show()
```
