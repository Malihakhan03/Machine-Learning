```python
#Remove the setosa category from dataseet for binary classification
df=df[df['species']!='setosa']
```

```python
df.species.unique()
```

```python
df.shape
```

```python
df = df.copy()
df['species'] = df['species'].map({'versicolor':0,'virginica':1})
```

```python
df.head()
```

```python
#split dataset into independent and dependent features
x=df.iloc[:,:-1]
print(x)
```

```python
y=df.iloc[:,-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=42)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
```

```python
#Hyper parameter tunning
from sklearn.model_selection import GridSearchCV
parameters={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300,400]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
#print the best parameters out of all taken
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
#prediction
y_pred= classifier_regressor.predict(x_test)
```

```python
#Accuracy Score
from sklearn.metrics import accuracy_score, classification_report
```

```python
score= accuracy_score(y_pred,y_test)
print(score)
```

```python
#for precision,Recall,F1 Score & Accuracy
print(classification_report(y_pred,y_test))
```
